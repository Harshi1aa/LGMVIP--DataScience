TASK : Image to Pencil Sketch with Python

Python views a picture as essentially an array of integers. In order to obtain some really intriguing findings, we may thus undertake a number of matrix modifications. With the help of this code, you can turn any image into a "pencil" outline. 

The image must first be read in RBG format before being converted to grayscale. A picture will become a traditional black and white photograph as a result. The grayscale picture, also known as the negative image, must then be inverted to create the inverted grayscale image. Details can be improved using inversion. Finally, by combining the grayscale image with the reversed blurry image, we can get the pencil sketch. The grayscale picture may be divided by the inverted fuzzy image to get this. As photos are essentially arrays, we can easily accomplish this programmatically using Python's division function.

**Steps Followed :**

1.  Read the input image as gray scale image
2.  Use a gaussian filter to eliminate any noise.
3.  Make a negative image (invert the image)
4.  Avoid these two pictures (gray and inverted images). Certain edges are strengthened by these two photos, while others are complemented.  
Alternately, these two pictures may be combined using various weights.
5.  Display the image and wait for the keystroke to shut down the display







```python
#Importing Required Library
import cv2
from google.colab.patches import cv2_imshow
```

**Importing Image**

```python
# Real image
image=cv2.imread("/content/download.jpg")
cv2_imshow(image)
```

**convert and create a new image in grayscale**

```python
grayscale_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) 
cv2_imshow(grayscale_image)
```

**Inverting the grayscale image**

```python
inverted_image = cv2.bitwise_not(grayscale_image)
cv2_imshow(inverted_image)
```

**Image Smoothing**

Blur the inverted image using GaussianBlur() function

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(blurred_image)
```

**Pencil sketch**

```python
def final_sketch(x,y):
  return cv2.divide(x, 255 - y, scale=256)     
```

```python
final_pencil_sketch = final_sketch(grayscale_image,blurred_image)
cv2_imshow(final_pencil_sketch)
```

**Image to Pencil Sketch with Python is done successfully!!**
